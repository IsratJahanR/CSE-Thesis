```python
import pandas as pd

data = pd.read_csv("clean_data.csv")

all_jobs = data["Prefer Job"].str.split(",").explode().str.strip()

jobs = all_jobs[all_jobs != ""].unique().tolist()

for job in jobs:
    data[job] = data['Prefer Job'].apply(lambda x: is_job_preferred(x, job))

total_students = len(data)

# List to store jobs with less than 5% preference
low_preference_jobs = []

# Calculate the percentage of students preferring each job and find jobs with less than 5%
for job in jobs:
    job_count = data[job].sum()  # Count of students preferring the job
    job_percentage = (job_count / total_students) * 100  # Calculate percentage
    
    if job_percentage < 5:  # Check if percentage is less than 5
        low_preference_jobs.append(job)  # Add job to the list
    
# Remove low-preference jobs from the 'Prefer Job' column
def remove_low_preference_jobs1(prefer_job_entry):
    for job in low_preference_jobs:
        prefer_job_entry = prefer_job_entry.replace(job, "").strip()  # Remove job from the string
    return " ".join(prefer_job_entry.split())  # Clean up extra spaces

def remove_low_preference_jobs(prefer_job_entry):
    if pd.isna(prefer_job_entry):
        return ""
    for job in low_preference_jobs:
        prefer_job_entry = prefer_job_entry.replace(job, "").strip()
    # Remove any trailing, leading, or multiple commas
    prefer_job_entry = ','.join([j.strip() for j in prefer_job_entry.split(",") if j.strip() != ""])
    return prefer_job_entry

# Apply the function to the 'Prefer Job' column
data['Prefer Job'] = data['Prefer Job'].apply(remove_low_preference_jobs)
# Drop the boolean columns created for each job
data = data.drop(columns=jobs)

columns_to_drop = ['Unnamed: 0']
data = data.drop(columns_to_drop, axis=1)
data.sample(10)
#data.to_csv('clean_data.csv')
#data.to_csv('tmp.csv')
university = data.groupby('Prefer Job').size().reset_index(name='Count')
university = university.sort_values(by='Count',ascending=False)
university.to_csv('tmp.csv')
```
